# üí° Data Preparation

I need to modify the data a little bit so if won't affect my analysis later when need

### Data Representation

 *I need the work on the "Month" column and change the from being represented with number to actual months of the year, that way it will have better understanding when we are using the column for analysis"

```python
month_mapping = {
    1: 'Jan',
    2: 'Feb',
    3: 'Mar',
    4: 'Apr',
    5: 'May',
    6: 'Jun',
    7: 'Jul',
    8: 'Aug',
    9: 'Sep',
    10: 'Oct',
    11: 'Nov',
    12: 'Dec'
}
```

```python
dff['Month'] = pd.to_numeric(dff['Month'], errors='coerce')
dff=dff[dff['Month'].between(1, 12)]  # filter out invalid months
dff['Month'] = dff['Month'].astype(int).map(month_mapping)
dff 
```
